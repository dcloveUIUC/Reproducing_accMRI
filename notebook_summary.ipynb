### Reproduction Summary:


The claim we are aiming to reproduce is the central idea that deep learning can be used to identify corruption within MRI scan images. Specifically, we reproduce the paper’s self-supervised k-space data corruption detector and show that it successfully flags signal data adversely affected by simulated patient motion and environmental noise. Due to computational limitations, we will begin by utilizing the single-coil knee data (vs. the multi-coil data highlighted n the paper) and attempt to attain similar results. To do so, we will adopt a similar model: a convolutional neural network that ideally outperforms the E2E-VarNet which was state-of-the-art prior to this paper.

Given the specific nature of this domain, this means that hospitals could widely implement this network and reduce MRI scan times, which would have a multitudinous and significant impact. If hospitals could reduce scan times, it would help deliver more reliable MRI results for children and other patients that cannot stay still, it would reduce the radiation patients receive by replacing some X-ray scans, increase the number of patients that can be evaluated, and make MRIs more viable for emergency situations that require quick responses.


### Methodology Explanation and Examples:


In attempting to reproduce the corruption detector, we utilized both code provided by the original author as well as other sources to create a working model. Those resources include the GitHub provided by the author (https://github.com/michael-s-yao/accMRI), Facebook Research’s repository (https://github.com/facebookresearch/fastMRI), and various useful tutorials that utilize Google Colab specifically (Fast_MRI_GAN, Fast_MRI). Due to a lack of funding for cloud resources, we utilize a home-grown gaming computer equipped with an NVIDIA 3070 ti GPU.
